    <h1>Platformer Player</h1>

    <p class="lead">
      <code>PlatformerPlayer</code> extends the base <code>Player</code> class to add
      <strong>gravity</strong>, <strong>jumping</strong>, and platformer-style
      <strong>collision resolution</strong> against <code>Barrier</code> objects &mdash; giving
      your character the physics it needs to walk, fall, and land on platforms.
    </p>

    <hr />

    <h2>Gravity System</h2>

    <p>
      Gravity is simulated by maintaining a <code>verticalVelocity</code> value that is
      decremented each frame, causing the player to accelerate downward continuously.
    </p>

    <!-- ── Section 1 ── -->
    <h3><span class="num">1</span> Key Properties</h3>

	<pre><span class="lang-tag">js</span><code>
const playerData = {
	id: 'Mario',

    <!-- ── Section 2 ── -->
    <h3><span class="num">2</span> Per-Frame Gravity Loop</h3>

    <p>
      Each frame, <code>update()</code> runs this sequence:
    </p>

    <pre><span class="lang-tag">js</span><code>// 1. Subtract gravity from vertical velocity (pulls player down)
this.verticalVelocity -= this.gravityAcceleration;
// 2. Convert to engine velocity (flip sign)
this.velocity.y = -this.verticalVelocity;
// 3. Move, draw, and check collisions
super.move();
super.draw();
super.collisionChecks();
// 4. After collisions, update grounded state
this.isGrounded = this._groundedThisFrame;</code></pre>

    <div class="callout">
      Gravity keeps subtracting every frame, so the player accelerates downward continuously
      unless a collision handler sets <strong><code>_skipGravityThisFrame = true</code></strong>.
    </div>

    <!-- ── Section 3 ── -->
    <h3><span class="num">3</span> Jumping</h3>

    <p>
      In <code>updateVelocity()</code>, a jump is triggered when the up key is pressed
      <strong>and</strong> the player is grounded (or has barrier support beneath them):
    </p>

    <pre><span class="lang-tag">js</span><code>if (upPressed && (this.isGrounded || this._hasBarrierSupport()) && !this._jumpPressedLatch) {
    this.verticalVelocity = this.jumpVelocity;  // Jump upward
    this.isGrounded = false;
    this._jumpPressedLatch = true;               // Prevent hold-to-jump
}</code></pre>

    <p>
      <code>_jumpPressedLatch</code> prevents the jump from re-triggering while the key is held.
      It resets when the key is released.
    </p>

    <hr />

    <!-- ── Section 4 ── -->
    <h3><span class="num">4</span> Detecting a Hit &mdash; <code>isCollision(other)</code></h3>

    <p>
      This overrides the base <code>Player</code> method. It computes two trimmed
      rectangles &mdash; one for the player, one for <code>other</code> &mdash; and checks
      if they overlap using a standard AABB test:
    </p>

    <pre><span class="lang-tag">js</span><code>const hit = (
    thisRect.left   < otherRect.right  &&
    thisRect.right  > otherRect.left   &&
    thisRect.top    < otherRect.bottom &&
    thisRect.bottom > otherRect.top
);</code></pre>

    <p>
      It also computes <strong>directional touch flags</strong> for both sides of the collision:
    </p>

    <pre><span class="lang-tag">js</span><code>touchPoints.this.top    // Player's top edge is touching the other's top
touchPoints.this.bottom // Player's bottom edge is touching the other's bottom
touchPoints.this.left   // ...and so on</code></pre>

    <div class="callout">
      These flags tell <strong><code>handleCollisionState()</code></strong> <em>how</em> the
      objects are touching, not just <em>that</em> they're touching.
    </div>

    <h2>Collision Resolution &mdash; <code>handleCollisionState()</code></h2>

    <p>
      Resolution behaves differently depending on whether the colliding object is a
      <code>Barrier</code>. Non-barrier collisions defer to <code>super.handleCollisionState()</code>.
      For <code>Barrier</code> objects, four cases are handled:
    </p>

    <!-- ── Section 5 ── -->
    <h3><span class="num">5</span> Barrier Cases</h3>

    <h4>1 &mdash; Landing on Top</h4>

    <pre><span class="lang-tag">js</span><code>if (touchPoints.top && this.verticalVelocity <= 0) {
    this.position.y = otherObject.y - (this.height - insets.bottom); // Snap to surface
    this.verticalVelocity = 0;
    this.isGrounded = true;
    this._skipGravityThisFrame = true;  // Don't re-apply gravity this frame
}</code></pre>

    <p>
      The guard <code>verticalVelocity &lt;= 0</code> is critical &mdash; it prevents this
      block from running on the same frame the player jumps, which would cancel the jump
      immediately.
    </p>

    <h4>2 &mdash; Hitting the Underside</h4>

    <pre><span class="lang-tag">js</span><code>if (touchPoints.bottom) {
    this.position.y = otherObject.y + otherObject.height - insets.top; // Push below
    this.verticalVelocity = 0;  // Kill upward momentum
}</code></pre>

    <h4>3 &mdash; Left Wall</h4>

    <pre><span class="lang-tag">js</span><code>if (touchPoints.left) {
    this.position.x = otherObject.x - (this.width - insets.right);
}</code></pre>

    <h4>4 &mdash; Right Wall</h4>

    <pre><span class="lang-tag">js</span><code>if (touchPoints.right) {
    this.position.x = otherObject.x + otherObject.width - insets.left;
}</code></pre>

    <h2>Changes to GameObject for preserving held inputs</h2>

    <!-- ── Section 6 ── -->
    <h3><span class="num">6</span>Why <code>clearPressedKeysOnCollision</code>exists</h3>

    <p>
      Platformer movement is continuous and the game reads held keys every frame to move the
      player. An earlier fix cleared <code>pressedKeys</code> before every collision pass to
      prevent stuck inputs caused by blocking dialogs (like <code>alert()</code>), which steal
      focus and prevent <code>keyup</code> from firing. This prevented clean movement in platformers, however, where holding a key is essential.
    </p>

    <div class="callout">
	  Setting <code>clearPressedKeysOnCollision = false</code> on an object opts it out of the
      key-clearing behavior, preserving held input through the collision pass. However, for
	  backwards compatability, if you don't set this property, the old key-clearing behavior will still run.

	  Breaking other games is fun but is unfortuately not the goal of this change
    </div>
