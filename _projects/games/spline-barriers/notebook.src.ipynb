    <h2>Spline Barriers</h2>

    <p class="lead">
      A <strong>SplineBarrier</strong> is a curved wall you can place anywhere in
      your scene. Unlike a normal rectangle Barrier, a SplineBarrier
      follows a smooth curve defined by a list of control points &mdash; letting you build slopes,
      arches, winding paths, and organic terrain boundaries.
    </p>

    <hr />

    <h2>How It Works</h2>

    <!-- ── Section 1 ── -->
    <h3><span class="num">1</span> Defining the Curve with Control Points</h3>

    <p>
      When you create a <code>SplineBarrier</code>, you pass in a <code>splinePoints</code>
      array &mdash; a list of <code>{ x, y }</code> coordinates that act as
      <strong>anchors</strong> along the path:
    </p>

    <pre><span class="lang-tag">js</span><code>const barrier = new SplineBarrier({
    id: 'cliff_edge',
    color: '#8B4513',
    lineWidth: 5,
    splinePoints: [
        { x: 50,  y: 400 },
        { x: 200, y: 300 },
        { x: 400, y: 350 },
        { x: 600, y: 200 },
    ]
}, gameEnv);</code></pre>

    <p>
      These points are <strong>not</strong> connected with straight lines. Instead, the barrier
      computes a smooth curve through them using <strong>Catmull-Rom spline interpolation</strong>.
    </p>

    <hr />

    <!-- ── Section 2 ── -->
    <h3><span class="num">2</span> Catmull-Rom Interpolation</h3>

    <p>
      The magic behind the smooth curve is the <code>catmullRom()</code> static method. It
      generates a point on the curve between two anchors (<code>P1</code> &rarr; <code>P2</code>),
      using their neighbors (<code>P0</code>, <code>P3</code>) as tension guides:
    </p>

    <pre><code>t = 0.0  →  start of segment (at P1)
t = 1.0  →  end of segment   (at P2)</code></pre>

    <p>
      By sampling this function many times per segment (default: 50 samples),
      <code>getCurvePoints()</code> builds a dense array of positions that forms the rendered and
      collidable curve. The more segments, the smoother and more collision-accurate the barrier.
    </p>

    <div class="callout">
      <strong></strong> You define <em>anchors (ie key points)</em> and the engine fills in the curve
      between them automatically!!
    </div>

    <hr />

    <!-- ── Section 3 ── -->
    <h3><span class="num">3</span> Rendering</h3>

    <p>
      Each <code>SplineBarrier</code> owns its own <strong>dedicated <code>&lt;canvas&gt;</code>
      element</strong>, layered above the game background (<code>z-index: 15</code>). On every
      frame, <code>draw()</code> clears that canvas and redraws the full curve using the computed
      points.
    </p>

    <p>
      This is intentionally separate from the main game canvas so the barrier doesn&rsquo;t
      interfere with other rendering layers and can be toggled via <code>visible</code>. This is
      for debugging purposes &mdash; you can set <code>visible: false</code> to hide the curve
      while keeping its collision active.
    </p>

    <hr />

    <!-- ── Section 4 ── -->
    <h3><span class="num">4</span> Collision Detection</h3>

    <p>
      <code>SplineBarrier</code> <strong>does not use the engine&rsquo;s standard rectangular
      hitbox</strong>. Both <code>collisionChecks()</code> and <code>isCollision()</code> are
      overridden to do nothing &mdash; rectangular collision is irrelevant for a curved boundary.
    </p>

    <p>Instead, <code>update()</code> runs its own proximity check every frame:</p>

    <ol>
      <li>Finds the <code>Player</code> object in <code>gameEnv.gameObjects</code></li>
      <li>Gets the player&rsquo;s center position</li>
      <li>Iterates over all curve points</li>
      <li>If the player is within <strong>20 pixels</strong> of any curve point, a collision is triggered</li>
    </ol>

    <p>
      On collision, the player is <strong>pushed away</strong> from the nearest curve point along
      the vector between them:
    </p>

    <samp class="formula">push direction = normalize(playerCenter &minus; collisionPoint) &times; pushStrength</samp>

    <p>This gives the illusion of a solid wall without needing a polygon collider.</p>
