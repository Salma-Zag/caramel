## Key Concepts at a Glance

| Concept | What It Does | When to Use It | Example |
|---------|-------------|-----------------|---------|
| **Serial Computing** | One task completes before the next starts | When tasks are dependent or speedup doesn't matter | Processing a single image |
| **Parallel Computing** | Multiple cores on ONE machine work simultaneously | When tasks are independent and data fits in RAM | Rendering 16 video frames on an 8-core laptop |
| **Distributed Computing** | Many SEPARATE machines collaborate over a network | When data is too large for one machine or scale is critical | Training an AI model on 10,000 GPUs across data centers |

## Common Misconceptions

<div style="background: #1a1915; border-left: 3px solid #d4843a; padding: 1rem 1.5rem; margin: 1rem 0; border-radius: 4px;">
  <p><strong style="color: #d4843a;">"More cores = always faster"</strong></p>
  <p style="color: #c4bcb0;">❌ False. Tasks must be parallelizable. Some work (like updating a single counter) can't be sped up by adding cores. See Amdahl's Law.</p>
</div>

<div style="background: #1a1915; border-left: 3px solid #d4843a; padding: 1rem 1.5rem; margin: 1rem 0; border-radius: 4px;">
  <p><strong style="color: #d4843a;">"Distributed = always better than parallel"</strong></p>
  <p style="color: #c4bcb0;">❌ False. Distributed adds network overhead. Parallel is faster when all data fits on one machine. Use distributed only when you need the scale.</p>
</div>

<div style="background: #1a1915; border-left: 3px solid #d4843a; padding: 1rem 1.5rem; margin: 1rem 0; border-radius: 4px;">
  <p><strong style="color: #d4843a;">"If one node crashes, the whole system fails"</strong></p>
  <p style="color: #c4bcb0;">❌ False. Fault-tolerant distributed systems automatically reassign work to healthy nodes. The job keeps running.</p>
</div>

## Quick Reference: When to Use What

```
Is the data larger than 1 machine can hold?
  → YES: Use DISTRIBUTED computing
  → NO: Continue below

Do all tasks fit on 1 machine but take too long?
  → YES: Use PARALLEL computing (multiple cores)
  → NO: Use SERIAL computing (1 core is fine)

Are tasks fully independent?
  → YES: Parallelism will help
  → NO: Sequential parts will limit speedup
```
