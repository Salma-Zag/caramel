function typeCode(text, onDone) {

  el.classList.add('hg-cursor'); el.textContent = '';
  G.typing = true;
  var i = 0, speed = Math.max(10, Math.min(25, 1400 / text.length));
  function tick() {
    if (i < text.length) { el.textContent += text[i++]; typeTimer = setTimeout(tick, speed); }
    else { G.typing = false; el.classList.remove('hg-cursor'); if (onDone) onDone(); }

    if (onDone) onDone();
  };
  tick();
}

  updateHUD(); fb.style.display = 'block';
  var nb = $('hg-nextbtn');
  nb.innerHTML = (L.isFinal ? 'Face the boss' : 'Next program') + ' &nbsp;<em class="hg-kbd">ENTER</em>';
  nb.style.display = 'block';
  kbar('hg-kbar', K+'ENTER'+Kc+' or '+K+'SPACE'+Kc+' &nbsp;·&nbsp; continue');
}

function bossResult(pick) {
  var o = BOSS.outcomes[pick], fb = $('hg-fb');
  fb.className = 'bad';
  fb.innerHTML = '<strong>' + o.label + ' ' + o.detail + '</strong><br>The other answer would have failed too. Any answer you give, this program defeats.';
  fb.style.display = 'block';
  var nb = $('hg-nextbtn');
  nb.innerHTML = 'See what this proves &nbsp;<em class="hg-kbd">ENTER</em>';
  nb.style.display = 'block';
  kbar('hg-kbar', K+'ENTER'+Kc+' or '+K+'SPACE'+Kc+' &nbsp;·&nbsp; continue');
}

function hgNext() {
  if (G.boss) { endGame(); return; }
  G.level++;
  if (G.level >= LEVELS.length) { G.boss = true; renderLevel(BOSS); }
  else { renderLevel(LEVELS[G.level]); }
}

function endGame() {
  G.screen = 'end';
  $('hg-game-screen').style.display = 'none';
  $('hg-end-screen').style.display  = 'block';
  setProgress(100);

    $('hg-end-icon').textContent  = '💀';

    $('hg-end-cards').innerHTML =

      card('gray', 'TRY AGAIN',
        'Reach the boss on a perfect run. The boss is where\n' +

    return;
  }
  $('hg-end-icon').textContent  = '🧠';
  $('hg-end-title').textContent = "You couldn't beat the boss. Nobody can.";
  $('hg-end-score').textContent = 'Score: ' + G.score + ' — ' + (G.score >= 500 ? 'perfect run' : G.score >= 300 ? 'solid' : 'keep practicing');
  $('hg-end-cards').innerHTML =
    card('blue', 'WHY THE BOSS WON — TURING\'S PROOF',

function card(cls, title, body) {
  return '<div class="hg-end-card ' + cls + '"><div class="hg-end-card-title">' + title + '</div>' +
    '<div style="white-space:pre;font-size:11.5px;line-height:1.85">' + body + '</div></div>';
}

function hgRestart() {
  G = { screen: 'game', slide: 0, level: 0, lives: 3, score: 0, done: false, boss: false, typing: false, help: false };
  $('hg-help-overlay').className    = '';
  $('hg-end-screen').style.display  = 'none';
  $('hg-game-screen').style.display = 'block';
  updateHUD(); renderLevel(LEVELS[0]);
}

function updateHUD() {
  var h = '';
  for (var i = 0; i < 3; i++) h += i < G.lives ? '♥' : '♡';
  $('hg-lives').textContent = h;
  $('hg-score').textContent = G.score;
}

function hgToggleHelp() {
  G.help = !G.help;
  $('hg-help-overlay').className    = G.help ? 'open' : '';
  $('hg-helpbtn').style.borderColor = G.help ? '#89b4fa' : '';
  $('hg-helpbtn').style.color       = G.help ? '#89b4fa' : '';
}

document.addEventListener('keydown', function(e) {
  var key = e.key.toLowerCase();
  if (key === '?' || key === '/') { e.preventDefault(); hgToggleHelp(); return; }
  if (key === 'escape') { if (G.help) { e.preventDefault(); hgToggleHelp(); } return; }
  if (G.help) return;
  if (G.screen === 'start') {
    if (key === 'enter' || key === ' ') { e.preventDefault(); hgStart(); }
  } else if (G.screen === 'intro') {
    if (key === 'enter' || key === ' ') { e.preventDefault(); hgNextSlide(); }
  } else if (G.screen === 'game') {
    if (key === 'arrowleft' || key === 'arrowright' || key === ' ') e.preventDefault();
    if (G.typing) {
      if (key === ' ' || key === 'enter' || key === 'h' || key === 'l' || key === 'arrowleft' || key === 'arrowright') {
        if (G.skipTyping) G.skipTyping();
      }
      return;
    }
    if (!G.done) {
      if (key === 'h' || key === 'arrowleft')  { flash('hg-halt'); hgGuess('halts'); }
      if (key === 'l' || key === 'arrowright') { flash('hg-loop'); hgGuess('loops'); }
    } else {
      if (key === 'enter' || key === ' ') { flash('hg-nextbtn'); setTimeout(hgNext, 80); }
    }
  } else if (G.screen === 'end') {
    if (key === 'enter' || key === ' ' || key === 'r') { e.preventDefault(); hgRestart(); }
  }
});

window.hgStart      = hgStart;
window.hgGuess      = hgGuess;
window.hgNext       = hgNext;
window.hgNextSlide  = hgNextSlide;
window.hgRestart    = hgRestart;
window.hgToggleHelp = hgToggleHelp;

})();
