## Overview

The Code Hub level contains three terminals — Frontend, Backend, and Dataviz. Each terminal has a lesson panel students read first, followed by a Space Invaders mini-game that quizzes them on that material.

**Problem:** The original quiz questions tested general prior knowledge instead of the lesson content students had just reviewed. A student could read the lesson perfectly and still fail the quiz.

**Fix:** All 15 questions were rewritten to pull directly from each terminal's reference panel, so the quiz rewards students for reading the lesson.

**File changed:** `_projects/games/cs-pathway/levels/GameLevelCsPath1CodeHub.js`

## Before vs After

### Frontend Terminal

**Lesson covers:** Markdown Cheat Sheet (`**bold**`, `*italic*`, `- list`, `> blockquote`, `` `code` ``) and Key CSS Concepts (Selectors, Box Model, Flexbox, Transitions, Gradients)

| | Old Question | New Question |
|---|---|---|
| Q1 | Which tag creates the biggest heading? → `<h1>` | In Markdown, `**text**` makes text ___ → `bold` |
| Q2 | What does CSS stand for? → `Cascading Style` | In Markdown, `*text*` makes text ___ → `italic` |
| Q3 | HTML is used to add __ to a webpage → `structure` | In Markdown, `- item` creates a ___ → `list` |
| Q4 | Which tag creates a paragraph? → `<p>` | In Markdown, `> text` creates a ___ → `blockquote` |
| Q5 | Where does CSS usually go in HTML? → `<head>` | CSS `display:flex` is called ___ → `Flexbox` |

Old questions tested HTML document structure — not covered in the Frontend lesson at all.

### Backend Terminal

**Lesson covers:** REST Methods reference — POST (Create), GET (Read), PUT (Update), DELETE (Remove)

| | Old Question | New Question |
|---|---|---|
| Q1 | A server SENDS data to a ___ → `client` | POST is used to ___ a resource → `create` |
| Q2 | What does a URL stand for? → `web address` | GET is used to ___ data → `read` |
| Q3 | HTTP is used to ___ data on the web → `transfer` | PUT is used to ___ a resource → `update` |
| Q4 | A backend sends back a ___ → `response` | DELETE is used to ___ data → `remove` |
| Q5 | Data sent to a server is called a ___ → `request` | To fetch all records, use ___ → `GET` |

Old questions tested client/server concepts not shown in the Backend lesson. Also, "What does a URL stand for?" → `web address` was factually incorrect (URL stands for Uniform Resource Locator, not "web address").

### Dataviz Terminal

**Lesson covers:** Query & Pagination Patterns — `?industry=`, `?minSize=`, `?page=&size=`, Spring JPA, JPQL

| | Old Question | New Question |
|---|---|---|
| Q1 | Charts and graphs show ___ → `data` | `?industry=Software` ___ by industry → `filters` |
| Q2 | A bar chart compares things using ___ → `bars` | `?minSize=100` finds companies ___ 100 → `over` |
| Q3 | A pie chart shows parts of a ___ → `whole` | `?page=1&size=4` is called ___ → `pagination` |
| Q4 | What library is used for JS charts? → `Chart.js` | JPQL queries begin with ___ → `SELECT` |
| Q5 | A line chart is best for showing ___ → `trends` | WHERE in JPQL is used to ___ results → `filter` |

Old questions were about chart types — the Dataviz lesson teaches filtering, pagination, and JPQL queries, not chart types.

## Why This Matters

The Code Hub is designed as an **entry-level course evaluation** — many students' first real impression of CS. If students do the right thing (read the lesson) and still fail the quiz because the questions test unrelated material, that creates a bad first impression.

The rule going forward: **every Space Invaders question must have its answer visible in the lesson panel before the student ever plays the game.**
