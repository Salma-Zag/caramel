<script>
  const fields = [
    "eventName",
    "eventDate",
    "eventLocation",
    "theme",
    "photoCaption",
    "beforeEvent",
    "curiousAbout",
    "whatSaw",
    "interestingThing",
    "learnedOne",
    "surprised",
    "inspired",
    "personalConnection",
    "talkedTo",
    "interaction",
    "changedThinking",
    "csConnection",
    "finalReflection",
    "takeaway"
  ];

  const storageKey = "interactive_event_reflection_v2";
  const imageStorageKey = "interactive_event_reflection_image_v2";

  function getValue(id) {
    return document.getElementById(id).value.trim();
  }

  function saveReflection() {
    const data = {};
    fields.forEach(id => {
      data[id] = getValue(id);
    });

    localStorage.setItem(storageKey, JSON.stringify(data));
  }

  function loadReflection() {
    const saved = localStorage.getItem(storageKey);

    if (saved) {
      const data = JSON.parse(saved);
      fields.forEach(id => {
        if (data[id]) {
          document.getElementById(id).value = data[id];
        }
      });
    }

    const savedImage = localStorage.getItem(imageStorageKey);
    if (savedImage) {
      renderImagePreview(savedImage);
    }
  }

  function renderImagePreview(imageData) {
    document.getElementById("imagePreviewArea").innerHTML = `
      <img src="${imageData}" alt="Uploaded event photo">
    `;
  }

  document.getElementById("photoUpload").addEventListener("change", function(event) {
    const file = event.target.files[0];
    if (!file) return;

    const reader = new FileReader();

    reader.onload = function(e) {
      const imageData = e.target.result;
      localStorage.setItem(imageStorageKey, imageData);
      renderImagePreview(imageData);
    };

    reader.readAsDataURL(file);
  });

  fields.forEach(id => {
    document.getElementById(id).addEventListener("input", saveReflection);
  });

  function section(title, body) {
    return `
      <h2>${title}</h2>
      <p>${body || "No response yet."}</p>
    `;
  }

  function generatePreview() {
    const imageData = localStorage.getItem(imageStorageKey);

    const imageHTML = imageData
      ? `<img src="${imageData}" alt="Uploaded event photo">`
      : `<p>No image uploaded yet.</p>`;

    const html = `
      <h1>${getValue("eventName") || "Event Reflection"}</h1>

      <p><strong>Date:</strong> ${getValue("eventDate") || "No date entered."}</p>
      <p><strong>Location:</strong> ${getValue("eventLocation") || "No location entered."}</p>
      <p><strong>Extra Credit Theme:</strong> ${getValue("theme") || "No theme selected."}</p>

      <h2>Proof of Experience</h2>
      ${imageHTML}
      <p>${getValue("photoCaption") || "No caption entered."}</p>

      ${section("Before the Event", getValue("beforeEvent"))}
      ${section("Curiosity Before Attending", getValue("curiousAbout"))}
      ${section("What I Saw", getValue("whatSaw"))}
      ${section("Most Interesting Thing I Saw", getValue("interestingThing"))}
      ${section("What I Learned", getValue("learnedOne"))}
      ${section("What Surprised Me", getValue("surprised"))}
      ${section("What Inspired Me", getValue("inspired"))}
      ${section("Personal Connection", getValue("personalConnection"))}

      <h2>Interaction</h2>
      <p><strong>Person or group I learned from:</strong> ${getValue("talkedTo") || "No response yet."}</p>
      <p>${getValue("interaction") || "No response yet."}</p>

      ${section("Depth of Experience", getValue("changedThinking"))}
      ${section("Connection to CS or Future Goals", getValue("csConnection"))}
      ${section("Final Reflection", getValue("finalReflection"))}

      <h2>Biggest Takeaway</h2>
      <blockquote>${getValue("takeaway") || "No takeaway entered yet."}</blockquote>
    `;

    document.getElementById("pdfContent").innerHTML = html;
    document.getElementById("previewArea").style.display = "block";
  }

  function downloadPDF() {
    generatePreview();

    const eventTitle = getValue("eventName") || "event-reflection";

    const options = {
      margin: 0.5,
      filename: eventTitle.replaceAll(" ", "-").toLowerCase() + ".pdf",
      image: { type: "jpeg", quality: 0.98 },
      html2canvas: {
        scale: 2,
        backgroundColor: "#ffffff"
      },
      jsPDF: {
        unit: "in",
        format: "letter",
        orientation: "portrait"
      }
    };

    html2pdf().set(options).from(document.getElementById("pdfContent")).save();
  }

  function downloadMarkdown() {
    const imageData = localStorage.getItem(imageStorageKey);
    const eventTitle = getValue("eventName") || "Event Reflection";

    const markdown = `---
layout: post
title: "${eventTitle}"
description: "Extra credit event reflection"
type: tangibles
comments: true
---

# ${eventTitle}

**Date:** ${getValue("eventDate")}

**Location:** ${getValue("eventLocation")}

**Extra Credit Theme:** ${getValue("theme")}

## Proof of Experience

${imageData ? "Image was uploaded in the interactive version. Add the image file to your blog repository and link it here." : "No image uploaded."}

**Caption:**  
${getValue("photoCaption")}

## Before the Event

${getValue("beforeEvent")}

## Curiosity Before Attending

${getValue("curiousAbout")}

## What I Saw

${getValue("whatSaw")}

## Most Interesting Thing I Saw

${getValue("interestingThing")}

## What I Learned

${getValue("learnedOne")}

## What Surprised Me

${getValue("surprised")}

## What Inspired Me

${getValue("inspired")}

## Personal Connection

${getValue("personalConnection")}

## Interaction

**Person or group I learned from:** ${getValue("talkedTo")}

${getValue("interaction")}

## Depth of Experience

${getValue("changedThinking")}

## Connection to CS or Future Goals

${getValue("csConnection")}

## Final Reflection

${getValue("finalReflection")}

## Biggest Takeaway

> ${getValue("takeaway")}
`;

    const blob = new Blob([markdown], { type: "text/markdown" });
    const url = URL.createObjectURL(blob);

    const link = document.createElement("a");
    link.href = url;
    link.download = eventTitle.replaceAll(" ", "-").toLowerCase() + ".md";
    link.click();

    URL.revokeObjectURL(url);
  }

  function clearReflection() {
    if (!confirm("Clear all saved responses?")) return;

    localStorage.removeItem(storageKey);
    localStorage.removeItem(imageStorageKey);

    fields.forEach(id => {
      document.getElementById(id).value = "";
    });

    document.getElementById("photoUpload").value = "";
    document.getElementById("imagePreviewArea").innerHTML = "";
    document.getElementById("pdfContent").innerHTML = "";
    document.getElementById("previewArea").style.display = "none";
  }

  loadReflection();
</script>
