# What is a Simulation?
A **simulation** is a program that models a real-world process or phenomenon.

Simulations are used when the real thing is:
- **Too dangerous** (crash testing, nuclear reactions)
- **Too expensive** (weather modeling, space missions)
- **Too slow** (evolution, climate change over centuries)

Simulations are always **abstractions** — they leave out details, so results may not perfectly match reality.

---

## Why Simulation Results Differ from Reality
1. **Simplifying assumptions** — not every real-world factor is included
2. **Incomplete data** — wrong inputs lead to wrong outputs
3. **Randomness** — two runs of the same simulation can give different results
4. **Missing variables** — you can only model what you think to include

```python
%%html
<!-- UI_RUNNER: Simulation Street interactive festival planner -->
<div id="sim-street-game"></div>

<script>
(() => {
  const mount = document.getElementById('sim-street-game');
  if (!mount) return;

  mount.innerHTML = `
    <style>
      #sim-street-game {
        --bg: #07111f;
        --panel: rgba(9, 24, 44, 0.88);
        --panel-2: rgba(18, 38, 67, 0.92);
        --line: rgba(139, 180, 255, 0.22);
        --text: #edf5ff;
        --muted: #9db3d1;
        --accent: #4de2c5;
        --accent-2: #ffd166;
        --danger: #ff6b6b;
        --success: #7ae582;
        --shadow: 0 24px 60px rgba(0, 0, 0, 0.32);
        color: var(--text);
        font-family: "Trebuchet MS", "Avenir Next", sans-serif;
        background:
          radial-gradient(circle at top left, rgba(77, 226, 197, 0.18), transparent 32%),
          radial-gradient(circle at top right, rgba(255, 209, 102, 0.18), transparent 28%),
          linear-gradient(160deg, #06101d 0%, #0b1830 45%, #081423 100%);
        width: min(100%, 1400px);
        margin: 0 auto;
        border-radius: 28px;
        padding: clamp(16px, 2vw, 24px);
        box-shadow: var(--shadow);
        border: 1px solid rgba(146, 183, 255, 0.18);
        overflow: hidden;
      }
      #sim-street-game * { box-sizing: border-box; }
      #sim-street-game .sim-shell {
        display: grid;
        grid-template-columns: minmax(0, 360px) minmax(0, 1fr);
        gap: 18px;
        align-items: start;
      }
      #sim-street-game .panel {
        background: var(--panel);
        border: 1px solid var(--line);
        border-radius: 22px;
        padding: 18px;
        backdrop-filter: blur(12px);
      }
      #sim-street-game .hero {
        margin-bottom: 16px;
        display: flex;
        justify-content: space-between;
        gap: 16px;
        align-items: flex-start;
        flex-wrap: wrap;
      }
      #sim-street-game .eyebrow {
        text-transform: uppercase;
        letter-spacing: 0.16em;
        font-size: 0.74rem;
        color: var(--accent);
        margin-bottom: 8px;
      }
      #sim-street-game h3,
      #sim-street-game h4,
      #sim-street-game p {
        margin: 0;
      }
      #sim-street-game .hero h3 {
        font-size: clamp(1.8rem, 3vw, 2.5rem);
        line-height: 1.05;
        margin-bottom: 8px;
      }
      #sim-street-game .hero p {
        max-width: 58ch;
        color: var(--muted);
        line-height: 1.5;
      }
      #sim-street-game .scorebox {
        min-width: min(100%, 320px);
        flex: 1 1 280px;
        background: rgba(255, 255, 255, 0.04);
        border: 1px solid rgba(255, 255, 255, 0.08);
        border-radius: 18px;
        padding: 14px;
      }
      #sim-street-game .scorebox strong {
        display: block;
        font-size: 2.2rem;
        color: var(--accent-2);
        margin-top: 6px;
      }
      #sim-street-game .controls {
        display: grid;
        gap: 14px;
      }
      #sim-street-game .control-group {
        background: rgba(255, 255, 255, 0.03);
        border: 1px solid rgba(255, 255, 255, 0.07);
        border-radius: 18px;
        padding: 14px;
      }
      #sim-street-game .control-row {
        display: flex;
        justify-content: space-between;
        align-items: center;
        gap: 12px;
        margin-bottom: 8px;
      }
      #sim-street-game label {
        font-weight: 700;
        font-size: 0.95rem;
      }
      #sim-street-game .hint,
      #sim-street-game .small {
        color: var(--muted);
        font-size: 0.84rem;
        line-height: 1.4;
      }
      #sim-street-game input[type="range"] {
        width: 100%;
        accent-color: var(--accent);
      }
      #sim-street-game .pill {
        display: inline-flex;
        align-items: center;
        justify-content: center;
        padding: 0.35rem 0.7rem;
        border-radius: 999px;
        font-size: 0.84rem;
        border: 1px solid rgba(255, 255, 255, 0.08);
        background: rgba(255, 255, 255, 0.05);
      }
      #sim-street-game .buttons {
        display: flex;
        flex-wrap: wrap;
        gap: 10px;
        margin-top: 10px;
      }
      #sim-street-game button {
        border: 0;
        border-radius: 999px;
        padding: 0.8rem 1rem;
        font-weight: 800;
        cursor: pointer;
        transition: transform 120ms ease, opacity 120ms ease, box-shadow 120ms ease;
      }
      #sim-street-game button:hover {
        transform: translateY(-1px);
      }
      #sim-street-game .primary {
        background: linear-gradient(135deg, #4de2c5, #21b1ff);
        color: #04101d;
        box-shadow: 0 10px 24px rgba(33, 177, 255, 0.28);
      }
      #sim-street-game .secondary {
        background: rgba(255, 255, 255, 0.08);
        color: var(--text);
        border: 1px solid rgba(255, 255, 255, 0.1);
      }
      #sim-street-game .city {
        background: var(--panel-2);
        border: 1px solid var(--line);
        border-radius: 24px;
        overflow: hidden;
        min-height: clamp(280px, 34vw, 420px);
        position: relative;
      }
      #sim-street-game .skyline {
        padding: 18px 18px 0;
        display: flex;
        justify-content: space-between;
        align-items: flex-start;
        gap: 12px;
      }
      #sim-street-game .city-title {
        font-size: 1.1rem;
        font-weight: 800;
      }
      #sim-street-game .event-chip {
        font-size: 0.84rem;
        color: #06101d;
        background: var(--accent-2);
        border-radius: 999px;
        padding: 0.45rem 0.7rem;
        font-weight: 800;
      }
      #sim-street-game .street {
        position: relative;
        margin: 20px 16px 18px;
        height: clamp(166px, 24vw, 240px);
        border-radius: 22px;
        overflow: hidden;
        background:
          linear-gradient(180deg, rgba(164, 214, 255, 0.22), rgba(164, 214, 255, 0.03) 38%),
          linear-gradient(180deg, #204466 0%, #172f4c 46%, #102238 46%, #102238 100%);
      }
      #sim-street-game .lane {
        position: absolute;
        left: 0;
        right: 0;
        top: 58%;
        height: 4px;
        background: repeating-linear-gradient(90deg, rgba(255,255,255,0.5), rgba(255,255,255,0.5) 18px, transparent 18px, transparent 34px);
        opacity: 0.45;
      }
      #sim-street-game .booth {
        position: absolute;
        bottom: 18px;
        width: 58px;
        height: 72px;
        border-radius: 14px 14px 10px 10px;
        background: linear-gradient(180deg, #ffd166 0%, #f9b64e 100%);
        border: 2px solid rgba(8, 20, 35, 0.35);
        box-shadow: 0 12px 20px rgba(0, 0, 0, 0.18);
      }
      #sim-street-game .booth::before {
        content: "";
        position: absolute;
        top: -14px;
        left: -2px;
        right: -2px;
        height: 18px;
        border-radius: 12px 12px 0 0;
        background: repeating-linear-gradient(90deg, #ff6b6b, #ff6b6b 12px, #fff0d4 12px, #fff0d4 24px);
        border: 2px solid rgba(8, 20, 35, 0.35);
      }
      #sim-street-game .booth-count {
        position: absolute;
        inset: 0;
        display: flex;
        align-items: center;
        justify-content: center;
        font-weight: 900;
        color: #082138;
      }
      #sim-street-game .visitor-layer {
        position: absolute;
        inset: 0;
      }
      #sim-street-game .visitor {
        position: absolute;
        width: 10px;
        height: 10px;
        border-radius: 50%;
        background: #9ed8ff;
        box-shadow: 0 0 0 4px rgba(158, 216, 255, 0.12);
        transform: translate(-50%, -50%);
        transition: transform 900ms cubic-bezier(.2,.8,.2,1), background 400ms ease, box-shadow 400ms ease;
      }
      #sim-street-game .visitor.served {
        background: var(--success);
        box-shadow: 0 0 0 4px rgba(122, 229, 130, 0.14);
      }
      #sim-street-game .visitor.waiting {
        background: var(--accent-2);
        box-shadow: 0 0 0 4px rgba(255, 209, 102, 0.16);
      }
      #sim-street-game .visitor.left {
        background: var(--danger);
        box-shadow: 0 0 0 4px rgba(255, 107, 107, 0.15);
      }
      #sim-street-game .dashboard {
        margin-top: 16px;
        display: grid;
        grid-template-columns: repeat(4, minmax(0, 1fr));
        gap: 10px;
      }
      #sim-street-game .metric {
        background: rgba(255, 255, 255, 0.04);
        border: 1px solid rgba(255, 255, 255, 0.07);
        border-radius: 18px;
        padding: 12px;
      }
      #sim-street-game .metric strong {
        display: block;
        font-size: 1.4rem;
        margin-top: 6px;
      }
      #sim-street-game .meter {
        height: 14px;
        background: rgba(255,255,255,0.08);
        border-radius: 999px;
        overflow: hidden;
      }
      #sim-street-game .meter > span {
        display: block;
        height: 100%;
        border-radius: inherit;
        background: linear-gradient(90deg, #ff6b6b, #ffd166, #4de2c5);
        width: 0%;
        transition: width 400ms ease;
      }
      #sim-street-game .analysis {
        margin-top: 16px;
        display: grid;
        grid-template-columns: 1.1fr 0.9fr;
        gap: 14px;
      }
      #sim-street-game .log,
      #sim-street-game .missions {
        background: rgba(255, 255, 255, 0.03);
        border: 1px solid rgba(255, 255, 255, 0.07);
        border-radius: 18px;
        padding: 14px;
      }
      #sim-street-game ul {
        list-style: none;
        padding: 0;
        margin: 10px 0 0;
      }
      #sim-street-game li {
        padding: 10px 0;
        border-top: 1px solid rgba(255,255,255,0.07);
        color: var(--muted);
        line-height: 1.4;
      }
      #sim-street-game li:first-child { border-top: 0; padding-top: 0; }
      #sim-street-game .mission {
        padding: 12px;
        border-radius: 16px;
        border: 1px solid rgba(255,255,255,0.07);
        background: rgba(255,255,255,0.03);
        margin-top: 10px;
      }
      #sim-street-game .mission.done {
        border-color: rgba(122, 229, 130, 0.45);
        background: rgba(122, 229, 130, 0.08);
      }
      #sim-street-game .lesson-card {
        margin-top: 14px;
        padding: 12px;
        border-radius: 16px;
        background: rgba(77, 226, 197, 0.08);
        border: 1px solid rgba(77, 226, 197, 0.18);
        color: #d9fff7;
      }
      @media (max-width: 1280px) {
        #sim-street-game .hero {
          flex-direction: column;
        }
        #sim-street-game .scorebox {
          width: 100%;
        }
        #sim-street-game .sim-shell,
        #sim-street-game .analysis {
          grid-template-columns: 1fr;
        }
        #sim-street-game .dashboard {
          grid-template-columns: repeat(2, minmax(0, 1fr));
        }
      }
      @media (max-width: 860px) {
        #sim-street-game .skyline {
          flex-direction: column;
        }
        #sim-street-game .event-chip {
          align-self: flex-start;
        }
        #sim-street-game .buttons button {
          flex: 1 1 100%;
        }
      }
      @media (max-width: 640px) {
        #sim-street-game {
          border-radius: 20px;
          padding: 14px;
        }
        #sim-street-game .panel,
        #sim-street-game .control-group,
        #sim-street-game .metric,
        #sim-street-game .mission,
        #sim-street-game .log,
        #sim-street-game .missions,
        #sim-street-game .scorebox {
          padding: 12px;
        }
        #sim-street-game .sim-shell,
        #sim-street-game .analysis {
          grid-template-columns: 1fr;
        }
        #sim-street-game .dashboard {
          grid-template-columns: 1fr;
        }
      }
    </style>

    <div class="hero">
      <div>
        <div class="eyebrow">3.16 Simulations</div>
        <h3>Simulation Street</h3>
        <p>Build a festival plan, then stress-test it with randomness. One run shows a single possible day. Many runs reveal whether your model actually holds up.</p>
      </div>
    </div>

    <div class="sim-shell">
      <div class="controls panel">
        <div class="control-group">
          <div class="control-row">
            <label for="visitors">Expected visitors</label>
            <span class="pill" id="visitors-value">180</span>
          </div>
          <input id="visitors" type="range" min="80" max="320" step="10" value="180" />
          <div class="hint">Your model starts with an estimate, not certainty. The real crowd can still swing above or below it.</div>
        </div>

        <div class="control-group">
          <div class="control-row">
            <label for="rain">Rain chance</label>
            <span class="pill" id="rain-value">20%</span>
          </div>
          <input id="rain" type="range" min="0" max="70" step="5" value="20" />
          <div class="hint">Weather changes attendance and buying behavior. A simulation needs assumptions about uncertain events.</div>
        </div>

        <div class="control-group">
          <div class="control-row">
            <label for="booths">Booths open</label>
            <span class="pill" id="booths-value">4</span>
          </div>
          <input id="booths" type="range" min="2" max="8" step="1" value="4" />
          <div class="control-row" style="margin-top: 12px;">
            <label for="capacity">Servings per booth</label>
            <span class="pill" id="capacity-value">34</span>
          </div>
          <input id="capacity" type="range" min="18" max="60" step="2" value="34" />
          <div class="hint">More capacity can reduce shortages, but overbuilding creates waste and costs money.</div>
        </div>

        <div class="control-group">
          <div class="control-row">
            <label for="trials">Batch trials</label>
            <span class="pill" id="trials-value">40</span>
          </div>
          <input id="trials" type="range" min="5" max="400" step="5" value="40" />
          <div class="hint">Small samples jump around. Large samples usually give a steadier picture.</div>
          <div class="buttons">
            <button class="primary" id="run-day">Run One Day</button>
            <button class="secondary" id="run-batch">Run Simulation Batch</button>
            <button class="secondary" id="reset-game">Reset Progress</button>
          </div>
        </div>
      </div>

      <div class="panel">
        <div class="city">
          <div class="skyline">
            <div>
              <div class="city-title">Festival Avenue</div>
              <div class="small">Watch a single possible day unfold. Dots turn green when served, yellow when lines get long, and red when people leave hungry.</div>
            </div>
            <div class="event-chip" id="event-chip">Awaiting trial</div>
          </div>
          <div class="street" id="street">
            <div class="lane"></div>
            <div class="visitor-layer" id="visitor-layer"></div>
          </div>
        </div>

        <div class="dashboard">
          <div class="metric">
            <div class="small">Profit</div>
            <strong id="profit-metric">$0</strong>
          </div>
          <div class="metric">
            <div class="small">Average wait</div>
            <strong id="wait-metric">0 min</strong>
          </div>
          <div class="metric">
            <div class="small">Sold</div>
            <strong id="sold-metric">0</strong>
          </div>
          <div class="metric">
            <div class="small">Turned away</div>
            <strong id="left-metric">0</strong>
          </div>
        </div>

        <div style="margin-top: 16px;">
          <div class="small" style="margin-bottom: 8px;">Plan confidence</div>
          <div class="meter"><span id="confidence-bar"></span></div>
          <div class="small" id="confidence-text" style="margin-top: 8px;">Run a batch to estimate how often this plan survives randomness.</div>
        </div>
      </div>
    </div>
  `;

  const controls = {
    visitors: mount.querySelector('#visitors'),
    rain: mount.querySelector('#rain'),
    booths: mount.querySelector('#booths'),
    capacity: mount.querySelector('#capacity'),
    trials: mount.querySelector('#trials')
  };

  const labels = {
    visitors: mount.querySelector('#visitors-value'),
    rain: mount.querySelector('#rain-value'),
    booths: mount.querySelector('#booths-value'),
    capacity: mount.querySelector('#capacity-value'),
    trials: mount.querySelector('#trials-value')
  };

  const ui = {
    street: mount.querySelector('#street'),
    visitorLayer: mount.querySelector('#visitor-layer'),
    eventChip: mount.querySelector('#event-chip'),
    profit: mount.querySelector('#profit-metric'),
    wait: mount.querySelector('#wait-metric'),
    sold: mount.querySelector('#sold-metric'),
    left: mount.querySelector('#left-metric'),
    confidenceBar: mount.querySelector('#confidence-bar'),
    confidenceText: mount.querySelector('#confidence-text')
  };

  const buttons = {
    runDay: mount.querySelector('#run-day'),
    runBatch: mount.querySelector('#run-batch'),
    reset: mount.querySelector('#reset-game')
  };

  function getSettings() {
    return {
      visitors: Number(controls.visitors.value),
      rain: Number(controls.rain.value),
      booths: Number(controls.booths.value),
      capacity: Number(controls.capacity.value),
      trials: Number(controls.trials.value)
    };
  }

  function updateLabels() {
    const settings = getSettings();
    labels.visitors.textContent = settings.visitors;
    labels.rain.textContent = `${settings.rain}%`;
    labels.booths.textContent = settings.booths;
    labels.capacity.textContent = settings.capacity;
    labels.trials.textContent = settings.trials;
    renderBooths(settings.booths);
  }

  function randomRange(min, max) {
    return min + Math.random() * (max - min);
  }

  function simulateDay(settings) {
    const rainRoll = Math.random() < settings.rain / 100;
    const weatherImpact = rainRoll ? randomRange(0.48, 0.82) : randomRange(0.94, 1.2);
    const buzz = randomRange(0.86, 1.18);
    const attendance = Math.max(40, Math.round(settings.visitors * weatherImpact * buzz));
    const buyRateBase = rainRoll ? randomRange(0.48, 0.68) : randomRange(0.56, 0.82);
    const buyRate = Math.min(0.92, buyRateBase + (attendance > settings.visitors ? 0.04 : 0));
    const demand = Math.round(attendance * buyRate);
    const capacity = settings.booths * settings.capacity;
    const shortage = Math.max(0, demand - capacity);
    const waste = Math.max(0, capacity - demand);
    const servicePressure = demand / Math.max(capacity, 1);
    const wait = Math.max(1, Math.round((servicePressure * 7) + randomRange(0.5, 4.5) + (rainRoll ? 1.2 : 0)));
    const sold = Math.min(capacity, demand);
    const left = Math.max(0, Math.round(shortage * randomRange(0.45, 0.8)));
    const setupCost = settings.booths * 52;
    const wasteCost = Math.round(waste * 1.5);
    const lostGoodwill = Math.round(left * 3.2);
    const profit = Math.round((sold * 7.5) - setupCost - wasteCost - lostGoodwill);
    const success = profit >= 0 && wait <= 10 && left <= 24;

    let event = 'Steady crowd';
    if (rainRoll && attendance < settings.visitors * 0.75) event = 'Rain slowed the crowd';
    else if (!rainRoll && attendance > settings.visitors * 1.12) event = 'Influencer spike';
    else if (shortage > 0) event = 'Supply crunch';
    else if (waste > settings.capacity) event = 'Overstocked stalls';

    return {
      attendance,
      demand,
      sold,
      left,
      waste,
      wait,
      profit,
      success,
      event,
      rainRoll,
      capacity
    };
  }

  function summarizeBatch(settings) {
    const runs = [];
    for (let i = 0; i < settings.trials; i += 1) {
      runs.push(simulateDay(settings));
    }
    const total = runs.reduce((acc, run) => {
      acc.profit += run.profit;
      acc.wait += run.wait;
      acc.left += run.left;
      acc.waste += run.waste;
      acc.success += run.success ? 1 : 0;
      return acc;
    }, { profit: 0, wait: 0, left: 0, waste: 0, success: 0 });

    return {
      trials: settings.trials,
      avgProfit: Math.round(total.profit / runs.length),
      avgWait: Number((total.wait / runs.length).toFixed(1)),
      avgLeft: Number((total.left / runs.length).toFixed(1)),
      avgWaste: Number((total.waste / runs.length).toFixed(1)),
      successRate: Math.round((total.success / runs.length) * 100),
      spread: Math.max(...runs.map((run) => run.profit)) - Math.min(...runs.map((run) => run.profit))
    };
  }
  function renderBooths(count) {
    mount.querySelectorAll('.booth').forEach((el) => el.remove());
    const street = ui.street;
    const gap = 100 / (count + 1);
    for (let i = 0; i < count; i += 1) {
      const booth = document.createElement('div');
      booth.className = 'booth';
      booth.style.left = `calc(${gap * (i + 1)}% - 29px)`;
      booth.innerHTML = `<div class="booth-count">${i + 1}</div>`;
      street.appendChild(booth);
    }
  }

  function renderVisitors(result) {
    const totalDots = Math.min(48, Math.max(18, Math.round(result.attendance / 5)));
    const servedDots = Math.round((result.sold / result.attendance) * totalDots);
    const leftDots = Math.round((result.left / result.attendance) * totalDots);
    const waitingDots = Math.max(0, totalDots - servedDots - leftDots);

    ui.visitorLayer.innerHTML = '';
    const dots = [];
    for (let i = 0; i < totalDots; i += 1) {
      const dot = document.createElement('div');
      dot.className = 'visitor';
      dot.style.left = `${8 + (i % 12) * 7}%`;
      dot.style.top = `${18 + Math.floor(i / 12) * 16}%`;
      ui.visitorLayer.appendChild(dot);
      dots.push(dot);
    }

    requestAnimationFrame(() => {
      dots.forEach((dot, index) => {
        let stateClass = 'served';
        if (index >= servedDots && index < servedDots + waitingDots) stateClass = 'waiting';
        if (index >= servedDots + waitingDots) stateClass = 'left';
        dot.classList.add(stateClass);
        const finalX = stateClass === 'served' ? 48 + (index % 7) * 5 : stateClass === 'waiting' ? 26 + (index % 9) * 4 : 86 + (index % 4) * 2;
        const finalY = stateClass === 'served' ? 12 + (index % 3) * 8 : stateClass === 'waiting' ? -8 + (index % 3) * 8 : 34 + (index % 5) * 9;
        dot.style.transform = `translate(${finalX}px, ${finalY}px)`;
      });
    });
  }

  function updateDayMetrics(result) {
    ui.eventChip.textContent = result.event;
    ui.profit.textContent = `$${result.profit}`;
    ui.wait.textContent = `${result.wait} min`;
    ui.sold.textContent = result.sold;
    ui.left.textContent = result.left;
  }
  function handleRunDay() {
    const settings = getSettings();
    const result = simulateDay(settings);
    updateDayMetrics(result);
    renderVisitors(result);
    ui.confidenceText.textContent = 'You just saw one possible future. Run a batch to estimate how often that kind of day happens.';
  }

  function handleRunBatch() {
    const settings = getSettings();
    const summary = summarizeBatch(settings);
    ui.confidenceBar.style.width = `${summary.successRate}%`;
    ui.confidenceText.textContent = `${summary.successRate}% of ${summary.trials} trials stayed profitable, reasonably fast, and kept turn-aways low. Average profit was $${summary.avgProfit} with ${summary.avgWait} minutes of wait time.`;
  }

  function resetGame() {
    ui.confidenceBar.style.width = '0%';
    ui.confidenceText.textContent = 'Run a batch to estimate how often this plan survives randomness.';
    ui.eventChip.textContent = 'Awaiting trial';
    ui.profit.textContent = '$0';
    ui.wait.textContent = '0 min';
    ui.sold.textContent = '0';
    ui.left.textContent = '0';
    ui.visitorLayer.innerHTML = '';
  }

  Object.values(controls).forEach((control) => {
    control.addEventListener('input', updateLabels);
  });

  buttons.runDay.addEventListener('click', handleRunDay);
  buttons.runBatch.addEventListener('click', handleRunBatch);
  buttons.reset.addEventListener('click', resetGame);

  updateLabels();
})();
</script>
```
