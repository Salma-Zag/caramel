## What is an Undecidable Problem?
An **undecidable problem** is a problem where no algorithm can always give a correct yes or no answer **for every possible input**.

This does not mean computers are slow or not powerful enough — it means it is **mathematically impossible** to write such a program, no matter how good the programmer is.

Even if a problem is very difficult, it can still be **decidable**.  
**Undecidable means impossible, not just hard.**

---

## The Halting Problem
The most famous undecidable problem is the **halting problem**:

> Can you write a program that looks at any other program and its input, and correctly determines whether that program will eventually stop (halt) or run forever?

The answer is **no**. It has been proven that no such program can exist.

For example, a program with a loop might run forever—or it might stop depending on conditions. There is no general algorithm that can correctly decide this for all programs.

Think of it this way — if you tried to build that program, you could always construct a case that breaks it, causing a contradiction.

---

## Decidable vs Undecidable

| Type | Definition | Example |
|------|------------|---------|
| **Decidable** | An algorithm always gives a correct yes/no answer for every input | Is this number even? |
| **Undecidable** | No algorithm can always give a correct yes/no answer for every input | Will this program halt or run forever? |

---

## Why This Matters
- It shows that **some problems cannot be solved by algorithms**
- It helps us understand the **limits of computing**
- It explains why we **can’t automate every possible task**

```python
%%html
<!-- UI_RUNNER: AP practice quiz for undecidable problems -->
<div id="undecidable-ap-quiz"></div>

<script>
(() => {
  const mount = document.getElementById('undecidable-ap-quiz');
  if (!mount) return;

  mount.innerHTML = `
    <style>
      #undecidable-ap-quiz {
        --text: #f4f8ff;
        --muted: #c7d4e7;
        --line: rgba(255,255,255,0.08);
        --sky: #7dd3fc;
        --mint: #86efac;
        --gold: #fcd34d;
        --rose: #fb7185;
        color: var(--text);
        font-family: "Trebuchet MS", "Avenir Next", sans-serif;
        background:
          radial-gradient(circle at 10% 12%, rgba(125, 211, 252, 0.18), transparent 24%),
          linear-gradient(145deg, #08111d 0%, #111b2b 42%, #0a111a 100%);
        width: min(100%, 1080px);
        margin: 0 auto;
        padding: clamp(16px, 2vw, 24px);
        border-radius: 24px;
        border: 1px solid rgba(255,255,255,0.06);
        box-shadow: 0 24px 54px rgba(0,0,0,0.28);
      }
      #undecidable-ap-quiz * { box-sizing: border-box; }
      #undecidable-ap-quiz h3,
      #undecidable-ap-quiz h4,
      #undecidable-ap-quiz p { margin: 0; }
      #undecidable-ap-quiz .hero,
      #undecidable-ap-quiz .quiz-card,
      #undecidable-ap-quiz .score-card,
      #undecidable-ap-quiz .option,
      #undecidable-ap-quiz .result {
        background: rgba(255,255,255,0.04);
        border: 1px solid rgba(255,255,255,0.08);
        border-radius: 20px;
      }
      #undecidable-ap-quiz .hero {
        padding: 18px;
        margin-bottom: 16px;
        background: linear-gradient(140deg, rgba(125, 211, 252, 0.12), rgba(134, 239, 172, 0.05));
      }
      #undecidable-ap-quiz .eyebrow {
        text-transform: uppercase;
        letter-spacing: 0.16em;
        font-size: 0.76rem;
        color: var(--mint);
        margin-bottom: 8px;
      }
      #undecidable-ap-quiz .hero h3 {
        font-size: clamp(1.9rem, 3vw, 2.5rem);
        margin-bottom: 8px;
      }
      #undecidable-ap-quiz .hero p,
      #undecidable-ap-quiz .tiny {
        color: var(--muted);
        line-height: 1.45;
        font-size: 0.94rem;
      }
      #undecidable-ap-quiz .layout {
        display: grid;
        grid-template-columns: 1fr;
        gap: 16px;
      }
      #undecidable-ap-quiz .quiz-card,
      #undecidable-ap-quiz .score-card {
        padding: 16px;
      }
      #undecidable-ap-quiz .question-count {
        font-size: 0.84rem;
        color: var(--sky);
        margin-bottom: 10px;
      }
      #undecidable-ap-quiz .question-text {
        font-size: 1.08rem;
        line-height: 1.5;
        margin-bottom: 14px;
      }
      #undecidable-ap-quiz .options {
        display: grid;
        gap: 10px;
      }
      #undecidable-ap-quiz .option {
        width: 100%;
        text-align: left;
        padding: 12px 14px;
        cursor: pointer;
        color: var(--text);
        transition: transform 120ms ease, border-color 120ms ease, background 120ms ease;
      }
      #undecidable-ap-quiz .option:hover { transform: translateY(-1px); }
      #undecidable-ap-quiz .option.correct {
        border-color: rgba(134, 239, 172, 0.35);
        background: rgba(134, 239, 172, 0.12);
      }
      #undecidable-ap-quiz .option.wrong {
        border-color: rgba(251, 113, 133, 0.35);
        background: rgba(251, 113, 133, 0.12);
      }
      #undecidable-ap-quiz .option:disabled {
        opacity: 1;
        cursor: default;
      }
      #undecidable-ap-quiz .button-row {
        display: flex;
        flex-wrap: wrap;
        gap: 10px;
        margin-top: 14px;
      }
      #undecidable-ap-quiz button {
        border: 0;
        border-radius: 999px;
        padding: 0.8rem 1rem;
        font-weight: 800;
        cursor: pointer;
      }
      #undecidable-ap-quiz .primary {
        background: linear-gradient(135deg, #7dd3fc, #38bdf8);
        color: #05121d;
      }
      #undecidable-ap-quiz .secondary {
        background: rgba(255,255,255,0.08);
        color: var(--text);
        border: 1px solid rgba(255,255,255,0.08);
      }
      #undecidable-ap-quiz .result {
        margin-top: 14px;
        padding: 14px;
        line-height: 1.45;
      }
      #undecidable-ap-quiz .result.correct {
        background: rgba(134, 239, 172, 0.12);
        border-color: rgba(134, 239, 172, 0.28);
        color: #e5ffed;
      }
      #undecidable-ap-quiz .result.wrong {
        background: rgba(251, 113, 133, 0.14);
        border-color: rgba(251, 113, 133, 0.3);
        color: #ffe2e8;
      }
      #undecidable-ap-quiz .meter {
        height: 14px;
        border-radius: 999px;
        background: rgba(255,255,255,0.08);
        overflow: hidden;
        margin-top: 8px;
      }
      #undecidable-ap-quiz .meter > span {
        display: block;
        height: 100%;
        width: 0%;
        border-radius: inherit;
        transition: width 320ms ease;
        background: linear-gradient(90deg, #86efac, #7dd3fc);
      }
      @media (max-width: 860px) {
        #undecidable-ap-quiz .layout { grid-template-columns: 1fr; }
      }
    </style>
    <div class="layout">
      <div class="quiz-card">
        <div class="question-count" id="question-count">Question 1</div>
        <div class="question-text" id="question-text"></div>
        <div class="options" id="options"></div>
        <div class="button-row">
          <button class="secondary" id="reset-btn">Reset Quiz</button>
          <button class="primary" id="next-btn" disabled>Next Question</button>
        </div>
        <div class="result" id="result-box" style="display:none;"></div>
      </div>
    </div>
  `;

  const questions = [
    {
      prompt: 'Which statement best describes an undecidable problem?',
      options: [
        'A problem that takes a very long time to solve',
        'A problem for which no algorithm can always give the correct yes/no answer',
        'A problem that needs a supercomputer',
        'A problem that has not been solved yet'
      ],
      answer: 1,
      explanation: 'Undecidable means no algorithm can always correctly answer every possible case.'
    },
    {
      prompt: 'Why is the halting problem undecidable?',
      options: [
        'Because some programs are too large to read',
        'Because computers are not fast enough',
        'Because a self-referential program can force any perfect predictor into contradiction',
        'Because loops are random'
      ],
      answer: 2,
      explanation: 'The proof uses self-reference: a program can react to the predictor and break it.'
    },
    {
      prompt: 'Which of these is most like a decidable problem?',
      options: [
        'Will any program halt on any input?',
        'Is this number even?',
        'Can a perfect halting detector exist?',
        'Will this self-referential program stop?'
      ],
      answer: 1,
      explanation: 'Checking whether a number is even is a simple problem with a guaranteed correct algorithm.'
    },
    {
      prompt: 'If a machine correctly predicts many easy halting cases, what does that prove?',
      options: [
        'It is a perfect halting decider',
        'Undecidable problems do not exist',
        'It works for some special cases, not necessarily all cases',
        'It can solve any AP CSP problem'
      ],
      answer: 2,
      explanation: 'Special-case success does not prove a universal solution exists.'
    }
  ];

  const state = {
    index: 0,
    answered: 0,
    correct: 0,
    locked: false
  };

  const ui = {
    questionCount: mount.querySelector('#question-count'),
    questionText: mount.querySelector('#question-text'),
    options: mount.querySelector('#options'),
    resetBtn: mount.querySelector('#reset-btn'),
    nextBtn: mount.querySelector('#next-btn'),
    resultBox: mount.querySelector('#result-box'),  };

  function renderQuestion() {
    const q = questions[state.index];
    state.locked = false;
    ui.questionCount.textContent = `Question ${state.index + 1} of ${questions.length}`;
    ui.questionText.textContent = q.prompt;
    ui.options.innerHTML = '';
    ui.resultBox.style.display = 'none';
    ui.resultBox.className = 'result';
    ui.resultBox.textContent = '';
    ui.nextBtn.disabled = true;

    q.options.forEach((option, idx) => {
      const button = document.createElement('button');
      button.className = 'option';
      button.textContent = option;
      button.addEventListener('click', () => answerQuestion(idx));
      ui.options.appendChild(button);
    });

  }

  function answerQuestion(choice) {
    if (state.locked) return;
    state.locked = true;
    state.answered += 1;
    const q = questions[state.index];
    const buttons = Array.from(ui.options.querySelectorAll('.option'));

    buttons.forEach((button, idx) => {
      button.disabled = true;
      if (idx === q.answer) button.classList.add('correct');
      if (idx === choice && idx !== q.answer) button.classList.add('wrong');
    });

    const isCorrect = choice === q.answer;
    if (isCorrect) state.correct += 1;

    ui.resultBox.style.display = 'block';
    ui.resultBox.className = `result ${isCorrect ? 'correct' : 'wrong'}`;
    ui.resultBox.textContent = `${isCorrect ? 'Correct.' : 'Not quite.'} ${q.explanation}`;
    ui.nextBtn.disabled = false;
    ui.nextBtn.textContent = state.index === questions.length - 1 ? 'Restart' : 'Next Question';

  }

  function nextQuestion() {
    if (state.index === questions.length - 1) {
      resetQuiz();
      return;
    }
    state.index += 1;
    renderQuestion();
  }

  function resetQuiz() {
    state.index = 0;
    state.answered = 0;
    state.correct = 0;
    state.locked = false;
    ui.nextBtn.textContent = 'Next Question';
    renderQuestion();
  }

  ui.resetBtn.addEventListener('click', resetQuiz);
  ui.nextBtn.addEventListener('click', nextQuestion);

  renderQuestion();
})();
</script>
```
