# Fault Tolerance of Networks

## What is Fault Tolerance?

Fault tolerance is the ability of a system to continue functioning properly in the event of the failure of some of its components. In the context of networks, fault tolerance refers to the ability of a network to maintain connectivity and provide services even when some of its nodes or links fail.

## Important Vocabularies
- **Fault Tolerance**: The ability of a system to continue operating properly in the event of a failure of some of its components.
- **Single Point of Failure (SPOF)**: A component in a system that, if it fails, will stop the entire system from working.
- **Redundancy**: The inclusion of extra components that are not strictly necessary for functioning, in case of failure of other components.
- **Network Redundancy**: The practice of adding extra links, nodes, or paths in a network to ensure that there are alternative routes for data to travel in case of a failure.
- **Failover**: The process of switching to a redundant or standby system upon the failure of the currently active system.

## Why is Fault Tolerance Important?
Fault tolerance is crucial for ensuring the reliability and availability of networks. In a fault-tolerant network, if one component fails, the network can reroute traffic through alternative paths, preventing service disruption. This is especially important for critical applications such as online banking, healthcare systems, and communication networks.

## Identifying Single Points of Failure
A single point of failure (SPOF) is a component in a system that, if it fails, will stop the entire system from working. In a network, a SPOF could be a single router, switch, or link that is critical for maintaining connectivity. Identifying and eliminating SPOFs is essential for improving the fault tolerance of a network.

## Fault Tolerance and the Internet 
The internet was intentionally designed to be fault-tolerant to ensure global connectivity remains stable even if local sections go offline. Some of the key design principles that contribute to the fault tolerance of the internet include:
- **Packet Switching**: Data is broken into small packets that can travel along different redundant paths to reach the same destination.
- **Routing**: Routers automatically detect failures in a path and update their routing tables to send subsequent packets through alternate routes

## Trade-offs of Fault Tolerance
Building a fault-tolerant system is not always the best choice; engineers must evaluate specific trade-offs: 
- **Increased Costs**: More hardware (routers, servers, cables) and maintenance required.
- **Complexity**: Managing multiple paths and redundant systems is more difficult than a single-path network.
- **Performance**: Redundant paths may introduce latency or reduce overall performance if not properly managed.

<script>
(function() {
  const nodes = [
    { id: 0, x: 300, y: 60,  label: 'R1' },
    { id: 1, x: 540, y: 200, label: 'R2' },
    { id: 2, x: 450, y: 385, label: 'R3' },
    { id: 3, x: 150, y: 385, label: 'R4' },
    { id: 4, x:  60, y: 200, label: 'R5' },
  ];

  const edges = [
    { id: 0, a: 0, b: 1 }, { id: 1, a: 1, b: 2 },
    { id: 2, a: 2, b: 3 }, { id: 3, a: 3, b: 4 },
    { id: 4, a: 4, b: 0 }, { id: 5, a: 0, b: 2 },
    { id: 6, a: 1, b: 3 }, { id: 7, a: 0, b: 3 },
    { id: 8, a: 1, b: 4 },
  ];

  const failedNodes = new Set();
  const failedEdges = new Set();
  let source = 0, dest = 2;

  function findPath(src, dst) {
    if (failedNodes.has(src) || failedNodes.has(dst)) return null;
    if (src === dst) return { nodes: [src], edges: [] };
    const prev = new Map();
    const visited = new Set([src]);
    const queue = [src];
    while (queue.length) {
      const cur = queue.shift();
      for (const e of edges) {
        if (failedEdges.has(e.id)) continue;
        const nb = e.a === cur ? e.b : e.b === cur ? e.a : null;
        if (nb === null || failedNodes.has(nb) || visited.has(nb)) continue;
        prev.set(nb, { from: cur, edge: e.id });
        if (nb === dst) {
          const pathNodes = [nb], pathEdges = [];
          let at = nb;
          while (prev.has(at)) { const p = prev.get(at); pathEdges.push(p.edge); at = p.from; pathNodes.unshift(at); }
          return { nodes: pathNodes, edges: pathEdges };
        }
        visited.add(nb); queue.push(nb);
      }
    }
    return null;
  }

  function ns(tag) { return document.createElementNS('http://www.w3.org/2000/svg', tag); }

  function render() {
    const svg = document.getElementById('ft-svg');
    svg.innerHTML = '';
    const result = findPath(source, dest);
    const pathEdgeSet = new Set(result ? result.edges : []);
    const pathNodeSet = new Set(result ? result.nodes : []);

    for (const e of edges) {
      const a = nodes[e.a], b = nodes[e.b];
      const failed = failedEdges.has(e.id);
      const onPath = pathEdgeSet.has(e.id);

      const hit = ns('line');
      hit.setAttribute('x1', a.x); hit.setAttribute('y1', a.y);
      hit.setAttribute('x2', b.x); hit.setAttribute('y2', b.y);
      hit.setAttribute('stroke', 'transparent'); hit.setAttribute('stroke-width', 18);
      hit.style.cursor = 'pointer';
      hit.addEventListener('click', () => { failedEdges.has(e.id) ? failedEdges.delete(e.id) : failedEdges.add(e.id); render(); });
      svg.appendChild(hit);

      const line = ns('line');
      line.setAttribute('x1', a.x); line.setAttribute('y1', a.y);
      line.setAttribute('x2', b.x); line.setAttribute('y2', b.y);
      line.setAttribute('stroke', failed ? '#ef4444' : onPath ? '#22c55e' : '#334155');
      line.setAttribute('stroke-width', onPath ? 5 : 2);
      if (failed) line.setAttribute('stroke-dasharray', '8 4');
      line.style.pointerEvents = 'none';
      svg.appendChild(line);

      if (failed) {
        const mx = (a.x+b.x)/2, my = (a.y+b.y)/2;
        const t = ns('text');
        t.setAttribute('x', mx); t.setAttribute('y', my+5);
        t.setAttribute('text-anchor','middle'); t.setAttribute('fill','#ef4444'); t.setAttribute('font-size','13');
        t.textContent = '✕'; t.style.pointerEvents = 'none';
        svg.appendChild(t);
      }
    }

    for (const n of nodes) {
      const failed = failedNodes.has(n.id);
      const isSrc = n.id === source, isDst = n.id === dest;
      const onPath = pathNodeSet.has(n.id);
      const g = ns('g');
      g.style.cursor = 'pointer';
      g.addEventListener('click', () => { failedNodes.has(n.id) ? failedNodes.delete(n.id) : failedNodes.add(n.id); render(); });

      const c = ns('circle');
      c.setAttribute('cx', n.x); c.setAttribute('cy', n.y); c.setAttribute('r', 30);
      c.setAttribute('fill',   failed ? '#450a0a' : isSrc ? '#14532d' : isDst ? '#3b0764' : onPath ? '#1e3a5f' : '#1e293b');
      c.setAttribute('stroke', failed ? '#ef4444' : isSrc ? '#22c55e' : isDst ? '#a855f7' : onPath ? '#3b82f6' : '#475569');
      c.setAttribute('stroke-width', (isSrc || isDst || onPath) ? 3 : 2);
      g.appendChild(c);

      const lbl = ns('text');
      lbl.setAttribute('x', n.x); lbl.setAttribute('y', n.y + (isSrc||isDst ? 1 : 5));
      lbl.setAttribute('text-anchor','middle');
      lbl.setAttribute('fill', failed ? '#fca5a5' : isSrc ? '#86efac' : isDst ? '#d8b4fe' : '#cbd5e1');
      lbl.setAttribute('font-size','14'); lbl.setAttribute('font-weight','bold');
      lbl.textContent = failed ? '✕' : n.label;
      g.appendChild(lbl);

      if (!failed && (isSrc || isDst)) {
        const badge = ns('text');
        badge.setAttribute('x', n.x); badge.setAttribute('y', n.y + 14);
        badge.setAttribute('text-anchor','middle');
        badge.setAttribute('fill', isSrc ? '#86efac' : '#d8b4fe');
        badge.setAttribute('font-size','9'); badge.setAttribute('font-weight','bold');
        badge.textContent = isSrc ? 'SRC' : 'DST';
        g.appendChild(badge);
      }

      svg.appendChild(g);
    }

    const status = document.getElementById('ft-status');
    if (source === dest) {
      status.textContent = 'Source and destination are the same';
      status.style.cssText += '; background:#1e293b; color:#94a3b8;';
    } else if (result) {
      const pathStr = result.nodes.map(id => nodes[id].label).join(' → ');
      status.innerHTML = `✅ Path: <strong>${pathStr}</strong>`;
      status.style.background = '#14532d'; status.style.color = '#86efac';
    } else {
      const s = nodes[source].label, d = nodes[dest].label;
      status.innerHTML = `❌ No path from <strong>${s}</strong> to <strong>${d}</strong> — partitioned`;
      status.style.background = '#450a0a'; status.style.color = '#fca5a5';
    }
  }

  const srcSel = document.getElementById('ft-src');
  const dstSel = document.getElementById('ft-dst');
  nodes.forEach(n => {
    [srcSel, dstSel].forEach(sel => {
      const opt = document.createElement('option');
      opt.value = n.id; opt.textContent = n.label; sel.appendChild(opt);
    });
  });
  srcSel.value = source; dstSel.value = dest;
  srcSel.addEventListener('change', e => { source = +e.target.value; render(); });
  dstSel.addEventListener('change', e => { dest   = +e.target.value; render(); });

  window.ftReset = function() { failedNodes.clear(); failedEdges.clear(); render(); };
  render();
})();
</script>

## TL;DR
- Fault tolerance is critical for network reliability and availability.
- Single points of failure can cause complete network outages.
- Redundancy and failover mechanisms help maintain connectivity during failures.
- Designing fault-tolerant networks involves trade-offs between cost, complexity, and reliability.
- Some trade-offs include increased costs, greater complexity, and potential scalability issues, but they provide essential resilience for critical applications.
